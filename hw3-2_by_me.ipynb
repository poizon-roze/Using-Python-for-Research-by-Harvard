```python
import os
import pandas as pd
import numpy as np
from collections import Counter

def count_words_fast(text):
    text = text.lower()
    skips = [".", ",", ";", ":", "'", '"', "\n", "!", "?", "(", ")"]
    for ch in skips:
        text = text.replace(ch, "")
    word_counts = Counter(text.split(" "))
    return word_counts

def word_stats(word_counts):
    num_unique = len(word_counts)
    counts = word_counts.values()
    return (num_unique, counts)
#hfkjhgkjherukgrhkdfjhgkjdfhgkjhfilkjdsfkjsdhgjdfhgikjdflbkjfgkjhnglfkjfgiljdlfkjblidfkjblfdk



f = 'https://courses.edx.org/asset-v1:HarvardX+PH526x+2T2019+type@asset+block@hamlets.csv'
hamlets = pd.read_csv(f, index_col = 0)
#hfkjhgkjherukgrhkdfjhgkjdfhgkjhfilkjdsfkjsdhgjdfhgikjdflbkjfgkjhnglfkjfgiljdlfkjblidfkjblfdk



language, text = hamlets.iloc[0]

counted_text = count_words_fast(text)
data = pd.DataFrame(columns = ('word', 'count'))

id = 1

print(counted_text['hamlet'])

for i in counted_text.keys():
    data.loc[id] = i, counted_text[i]
    id += 1
#hfkjhgkjherukgrhkdfjhgkjdfhgkjhfilkjdsfkjsdhgjdfhgikjdflbkjfgkjhnglfkjfgiljdlfkjblidfkjblfdk


word_len = []

for i in counted_text.keys():
    word_len.append(len(i))
    
data['length'] = word_len
data.head(20)


fr = []
for i in data['count']:
    if  i > 10:
        fr.append('frequent')
    elif i > 1 and i <= 10:
        fr.append('infrequent')
    else:
        fr.append('unique')

data['frequency'] = fr
data['frequency'].value_counts()
#hfkjhgkjherukgrhkdfjhgkjdfhgkjhfilkjdsfkjsdhgjdfhgikjdflbkjfgkjhnglfkjfgiljdlfkjblidfkjblfdk


mean_word_length = {}

new_dfu = data[data['frequency'] == 'unique']
mean_u = sum(new_dfu['length'])/len(new_dfu['length'])
mean_word_length['unique'] = mean_u

new_dff = data[data['frequency'] == 'frequent']
mean_f = sum(new_dff['length'])/len(new_dff['length'])
mean_word_length['frequent'] = mean_f

new_dfi = data[data['frequency'] == 'infrequent']
mean_i = sum(new_dfi['length'])/len(new_dfi['length'])
mean_word_length['infrequent'] = mean_i

num_words = {}

num_words['unique'] = len(new_dfu)
num_words['frequent'] = len(new_dff)
num_words['infrequent'] = len(new_dfi)

frequency = ['unique', 'frequent', 'infrequent']

sub_data = pd.DataFrame(columns = ('language', 'frequency', 'mean_word_length', 'num_words'))

sub_data['language'] = language
sub_data['frequency'] = frequency
sub_data['mean_word_length'] = mean_word_length.values()
sub_data['num_words'] = num_words.values()
#hfkjhgkjherukgrhkdfjhgkjdfhgkjhfilkjdsfkjsdhgjdfhgikjdflbkjfgkjhnglfkjfgiljdlfkjblidfkjblfdk



```
